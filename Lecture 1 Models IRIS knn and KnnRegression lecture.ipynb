## KNN Classification

In the simplist form the k-NN algo only considers exactly one nearest neighbor  with is the closest data point to the point we are looking at in the test data we want a prediction for. The prediction is then the known output for this training point.

Using the Forge synthetic dataset we can plot what happens:
- Added 3 new data points, seen as stars.
- marked their closest point in the training set.
- The prediction of the one-nearest neighbor algo is the label of that point.


```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

The k in k-Nearest Neighbor lets you choose ANY arbitrary number of neighbors. We use voting to assign the label. For each test point we count how many neighbors belong to class 0 and how many to class 1, then we assign the class that is more frequent. Lets plot this:

The predictions are shows as the color of the most frequent nearest neighbors. See the last plot and notice that the top left one is not the same as when we used only one nearest neighbor. 

In this case we are looking at forge dataset that is a binary classification, but this works for any number of classes. We count how many neighbors belong to each class and again predict the most common class. 

```python
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

Applying K-NN with Scikit-learn
- import the splitter function
- import the data (forge data here)
- split the data into a test and train set so we can evaluate the performance of our generalization.
- import and instantiate the class, and set parameters like the number of neighbors we want to use.
- fit the classifier using the taining set (in the case of the k-NN it really just means storing the data so we can compute the nearest predictions when we add the test data)
- use predict() to make predictions. For each data point in the test data we will compute the nearest neighbor in the training set and find the most common of those returned. 
- Evaluate how well our model generalizes by using the score(), with the test data (X) and test labels (y).
- Note that the model predicted the right class for the test samples at 86% accuracy for the forge dataset. 

```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)

clf.fit(X_train, y_train)
print("Test set predictions: {}".format(clf.predict(X_test)))
print("Test set accuracy: {:.2f}".format(clf.score(X_test, y_test)))
print("Test set accuracy in percentage: {:.2f}%".format(clf.score(X_test, y_test)*100))
```

## Analyzing the K-NN classifier

Rememeber we use a line for a single feature. When we have a small 2 feature dataset we can illustrate all possible predictions for all possible test points in the XY plane. We can also color the place according to the class that will be assigned to any point in that colored region. It allows us to create and see a **decision boundary** which is the divide between the classes.

We plot three graphs for different neighbor numbers to see the differences when you choose different k numbers. The left figure uses only k=1. and the boundary follows the training data very very closely. As we add more to the k the boundary gets smoother. **A smoother boundary corresponds to a simpler model** therefore using a lower k number (fewer neighbors) leads to a high model complexity. If we go to the extreme and make the k the same as ALL the training data points then each test point would all have the same neighbors and all predictions would be the same. Whichever class is the most frequent in all the training data. 


```python
fig, axes = plt.subplots(1, 4, figsize=(10, 4))
for n_neighbors, ax in zip([1, 3, 9, 6], axes):
    #the fit method returns the object self, so we can instantiate and fit in one line
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:,1], y, ax=ax)
    ax.set_title("{} neighbors".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
axes[0].legend(loc=4)
```

#### Can we confirm the connection between model complexity and generalization? 

Lets use the Breast Cancer dataset.
- split the dataset into a training and testing set
- evaluate the training and test performance with different numbers of neighbors

- y-axis is the training and test set accuracy
- x-axis is the setting we choose for the n_neighbor number
- considering fewer neighbors corresponds to a more complex model

If you watch the training data predictions:
- considering a k=1 the model is perfect on the training data's predictions.
- As you add more to the k (more neighbors are considered) the model becomes simpler and the training accuracy drops.

If you watch the test data predictions:
- The test set accuracy for using a k=1 is lower than when using more neighbors indicating that using the single nearest neighbor leads to a model that is too complex.
- When considering k=10 the model is too simple and performace is even worse.
- The best performace is somewhere around k=6.

Keep the scale of the plot in mind: The worst performace is around 88% accuracy. Which might still be acceptible depending on what we end up plotting. 

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []

# try n_neighbors form 1-10
neighbors_settings = range(1, 11)

for n_neighbors in neighbors_settings:
    #build the model
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    # record trainign set accuracy
    training_accuracy.append(clf.score(X_train, y_train))
    # record generalization accuracy
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.xlabel("n_neighbors")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Comparison of training and test accuracy as a function of n_neighbor")
```

#### R^2 Coefficients of Determination for Regression Models
Evaluate the model with the score method, in regressions this returns the coefficients of determination which is seen as R^2 score.
- Is a measure of goodness of a prediction for a regression model. 
- Yields a score between 1 and 0. 
- 1 = a perfect prediction 
- 0 = a constant model that just predicts the mean of the training set responses, y_train.


```python
from sklearn.neighbors import KNeighborsRegressor

X, y = mglearn.datasets.make_wave(n_samples=40)

#split the wave dataset into training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

#instantiate the model and set the number of neighbors to consider k=3
reg = KNeighborsRegressor(n_neighbors=3)
#fit the model using the training data and the training targets
reg.fit(X_train, y_train)

#make predictions on the test set
#note these are numbers see the score for the R^2 coefficients of determination
print("Test set predictions: \n{}".format(reg.predict(X_test)))

#Evaluate the model with the score method, in regressions this returns the coefficients of determination which is seen as R^2 score, and is a measure of goodness of a prediction for a regression model. and yields a score between 1 and 0. 1 = a perfect prediction, 0=a constant model that just predicts the mean of the training set responses, t_train.
print("Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))



```

Since we are using a 1-D dataset here in the FORGE synthetic dataset we can see what the predictions look like for all possible feature values.

We will create test data that have many points on a line. 
- Using a k=1 each point in the training set has an obvious influence on the predictions and the prediction values go through all the data points.
- This is very unsteady prediction,
- as you add more to the k value the predictions get smoother. But, as before, they don't fit the trainign data that well.

Comparing predictions made by nearest neighbors regression for different values of n_neighbors k

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
#create 1K data points, evenly spaced between -3 and 3
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
for n_neighbors, ax in zip([1, 3, 9], axes):
    #make predictions using 1, 3, 9 neighbors for the k values
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)
    
    ax.set_title(
    "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(
        n_neighbors, reg.score(X_train, y_train), 
        reg.score(X_test, y_test)))
    ax.set_xlabel("Feature")
    ax.set_label("Target")
axes[0].legend(["Model Predictions", "Training data/target", "Test data/target"], loc="best")
    
```

# Final words about K-NN

2 important parameters for the KNeighbors classifiers
- number of neighbors aka k. using a small number such as 3 or 5 works well in many situations. 
- distance to measure between points. But, we won't dive into this, it's Euclidian and works well in most setting. But, know it can be changed. 

- easy to understand
- gives resonable performance without much adjutment necessary
- a good baseline before trying more advanced things (See SVM and Naive Baise as this is a good benchmark for those)
- building the model is usually fast
- if the trainign set is very large in either features or samples prediction time will be slow
- you must preprocess your data (see how to process unfeatured data later on)
- does not work well on datasets with hundreds or more features
- does badly on datasets with sparse features (sparse features have lots of zeros in them, we will see others that do works well with this, also pandas has a special data object to hold extremely large sparse data sets.)
